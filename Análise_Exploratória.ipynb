```python
import dask.dataframe as dd

# Read the CSV file into a Dask DataFrame with explicit encoding specification
ddf = dd.read_csv('/content/drive/MyDrive/ENEM_DATA/MICRODADOS_ENEM_2021.csv', delimiter=';', encoding='latin1', dtype={'Q005': 'float64'})

# Check if any of the specified columns have missing values in a row and keep only those rows
ddf = ddf.dropna(subset=['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT'], how='all')

# Save the cleaned Dask DataFrame back to a CSV file
ddf.to_csv('cleaned_dataset.csv', index=False, sep=';', single_file=True, compute=True)
```
